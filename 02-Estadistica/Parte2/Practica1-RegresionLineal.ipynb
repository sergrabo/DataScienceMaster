## 1. Regresión lineal 

Comenzamos realizando una regresión lineal. La realizaremos de varias maneras diferentes: 
* A través de las expresiones analíticas de un modelo de regresión lineal con dos parámetros
* Mediante las expresiones generales de regresión en notación matricial
* Utilizando la función `lm()` implementada en R

### 1.1. Expresiones analíticas

Como hemos visto en la teoría, estas expresiones son:
$$ 
\alpha_0 = \overline{y}-\alpha_1\overline{x}
$$

$$
\alpha_1 = \frac{\overline{xy}-\bar{x}\bar{y}}{\overline{x^2}-\overline{x}^2}
$$
Así que implementaremos directamente estas expresiones en el código.

```r
alpha1 = (mean(x*y) - mean(x)*mean(y))/(mean(x*x) - mean(x)*mean(x))
alpha0 = mean(y) - alpha1*mean(x)
alphaLin = c(alpha0, alpha1)
```

```r
plot(x, y)
abline(alphaLin, col = 'red')
```

Se puede apreciar cierta linealidad en el ajuste, aunque no todos los puntos pasen por la recta.

### 1.2. Notación matricial

En notación matricial, la expresión que nos da los valores de los parámetros de ajuste es 
$$
\alpha = (X^TX)^{-1}X^Ty
$$
donde 
$$
     \alpha = \left(\begin{array}{c}
        \alpha_0\\
        \alpha_1\\
        \alpha_2\\
        \vdots\\
        \alpha_m
        \end{array}
     \right),\quad
     X = \left(\begin{array}{ccccc}
        1 & x_1^1 & x_2^1 & \cdots & x_n^1\\
        1 & x_1^2 & x_2^2 & \cdots & x_n^2\\
        1 & x_1^3 & x_2^3 & \cdots & x_n^3\\
        \vdots & \vdots & \vdots & \ddots & \vdots\\
        1 & x_1^m & x_2^m & \cdots & x_n^m
        \end{array}
     \right),\quad
     y = \left(\begin{array}{c}
        y^1\\
        y^2\\
        y^3\\
        \vdots\\
        y^n
        \end{array}
    \right)
$$
Definimos una función para realizar la regresión:

```r
# Función auxiliar para transformar los datos
# a notación matricial para polinomios de grado 1
transPol1=function(x){
  
  ones = rep(1, length(x))
  xMatrix = cbind(ones, x)
  return(xMatrix)
}

#Función para realizar la regresión lineal
ajusteRecta = function(x,y){
  
  xMatrix = transPol1(x)
  alpha = solve(t(xMatrix)%*%xMatrix) %*% t(xMatrix) %*% y
  return(alpha)
}
```

```r
alphaMat = ajusteRecta(x,y)
```

### 1.3. Función `lm()`

```r
alphaLm = lm(y ~ x)
```

Y finalmente comparamos los resultados obtenidos:

```r
dfLineal = data.frame(
            Alpha0 = c(alphaLin[1], alphaMat[1], alphaLm$coefficients[1]),
            Alpha1 = c(alphaLin[2], alphaMat[2], alphaLm$coefficients[2])
)
row.names(dfLineal) = c("Lineal", "Matricial", "Lm")
dfLineal
```

Como podemos observar, se obtienen parámetros exactamente iguales mediante los tres métodos. Comparando el ajuste lineal mediante las expresiones teóricas y el ajuste mediante notación matricial (primero y segundo), es normal que se obtenga el mismo resultado, ya que las expresiones analíticas se obtienen tras operar la ecuación matricial para el caso particular de dos parámetros (regresión lineal). Además, como ambos coinciden con los resultados obtenidos mediante `lm()`, podemos suponer que son correctos.

## 2. Regresión no lineal
Pasamos ahora a realizar regresiones no lineales, concretamente utilizando polinomios de segundo y tercer grado. Para construir la matriz $X$ de variables independientes, sabiendo que únicamente disponemos de un set de datos, simplemente añadiremos filas a $X$ donde multiplicamos las x's por si mismas el número de veces que sea necesario para obtener el grado del polinomio con el que deseamos realizar la regresión. Para ello, definimos una función similiar a la anterior para acomodar la matriz $X$, y posteriormente realizamos la regresión haciendo uso de la notación matricial.

### 2.1 Polinomio de segundo grado

```r
# Función auxiliar para transformar los datos
# a notación matricial para polinomios de grado 2
transPol2=function(x){
  
  ones = rep(1, length(x))
  xMatrix = cbind(ones, poly(x, 2, raw = T))
  return(xMatrix)
  
}
#Función para realizar la regresión
ajustePol2=function(x,y){
  
  xMatrix = transPol2(x)
  alpha = solve(t(xMatrix)%*%xMatrix) %*% t(xMatrix) %*% y
  return(alpha)
}
```

```r
alphaPol2 = ajustePol2(x,y)
```

Comparamos con los resultados que se obtienen mediante la función `glm()`

```r
alphaGlm2 = glm(y ~ poly(x, 2, raw = T))
```

```r
dfPol2 = data.frame(
            "Alpha0" = c(alphaPol2[1], alphaGlm2$coefficients[1]),
            "Alpha1" = c(alphaPol2[2], alphaGlm2$coefficients[2]),
            "Alpha2" = c(alphaPol2[3], alphaGlm2$coefficients[3])
)
row.names(dfPol2) = c("Matricial", "Glm")
dfPol2
```

En este caso, los parámetros que obtenemos son exactamente iguales, por lo que confirmamos que la regresión para el polinomio de segundo grado es correcta.

### 2.2 Polinomio de tercer grado

Repetimos el mismo proceso para una regresión con un polinomio de tercer grado.

```r
# Función auxiliar para transformar los datos
# a notación matricial para polinomios de grado 3
transPol3=function(x){
  
  ones = rep(1, length(x))
  xMatrix = cbind(ones, poly(x, 3, raw = T))
  return(xMatrix)
  
}
#Función para realizar la regresión
ajustePol3=function(x,y){
  
  xMatrix = transPol3(x)
  alpha = solve(t(xMatrix)%*%xMatrix) %*% t(xMatrix) %*% y
  return(alpha)
}
```

```r
alphaPol3 = ajustePol3(x,y)
```

Y comparamos los resultados con los proporcionados por `glm()`

```r
alphaGlm3 = glm(y ~ poly(x, 3, raw = T))
```

```r
dfPol3 = data.frame(
            "Alpha0" = c(alphaPol3[1], alphaGlm3$coefficients[1]),
            "Alpha1" = c(alphaPol3[2], alphaGlm3$coefficients[2]),
            "Alpha2" = c(alphaPol3[3], alphaGlm3$coefficients[3]),
            "Alpha3" = c(alphaPol3[4], alphaGlm3$coefficients[4])
)
row.names(dfPol3) = c("Matricial", "Glm")
dfPol3
```

Nuevamente, los resultados que obtenemos son exactamente iguales, por lo que la regresión es correcta.

## 3. Comparación de modelos a través de la función de Loss

Finalmente, vamos a comparar la validez de los tres modelos haciendo uso de la función de coste, o *loss function*. Para esta práctica en concreto la definimos como el cuadrado de la distancia euclídea, que en notación matricial viene dada como:
$$
\text{Loss} = (y-X\alpha)^T (y-X\alpha)
$$
La definimos a través de una función de R, cuyos parámetros de entrada serán $y$, $X$ y $\alpha$.

```r
loss = function(x, y, alpha){
  
  l = t(y-x%*%alpha) %*% (y-x%*%alpha)
  
}
```

Para realizar esta comparación, utilizaremos datos del año 2015, por lo que tendremos que volver a despurarlos para quedarnos con los relevantes

```r
# Como ya tenemos la fecha formateada, 
# tomamos directamente el subconjunto de 2015
df = subset(dataRaw, format(date, "%Y") == '2015')

#Quitamos los NA
curatedData2015 = drop_na(df)
```

```r
#Definimos las variables independientes y dependientes
x = curatedData2015$AVG.Cond
y = curatedData2015$AVG.Salinity
```

### 3.1. Ajuste lineal

```r
alpha1 = alphaLin
xMat = transPol1(x)
loss1 = loss(xMat, y, alpha1)
```

### 3.2. Polinomio grado 2

```r
alpha2 = alphaPol2
xMat = transPol2(x)
loss2 = loss(xMat, y, alpha2)
```

### 3.3. Polinomio grado 3

```r
alpha3 = alphaPol3
xMat = transPol3(x)
loss3 = loss(xMat, y, alpha3)
```

Ahora, simplemente comprobamos cuál de los tres valores de *loss* es menor.

```r
c(as.numeric(loss1), as.numeric(loss2), as.numeric(loss3))
```

```r
which.min(c(loss1, loss2, loss3))
```

Se observa que la regresión con menor función de coste es la del polinomio de grado 2, por lo que supone el mejor ajuste de todos los modelos que hemos planteado. De hecho, si ploteamos los ajustes:

```r
#Ordenamos los datos a plotear
ix <- sort(x,index.return=T)$ix

#Calculamos los puntos donde caen las predicciones
y1 = alphaLin[1] + x*alphaLin[2]
y2 = alphaPol2[1] + alphaPol2[2]*x + alphaPol2[3]*x^2 
y3 = alphaPol3[1] + alphaPol3[2]*x + alphaPol3[3]*x^2 + alphaPol3[4]*x^3

#Ploteamos los tres modelos
plot(x,y, pch=16, cex=0.7, main = "Regresión Lineal", xlab = "Conductividad", ylab = "Salinidad")
lines(x[ix], y1[ix], col = "red")
lines(x[ix], y2[ix], col = "blue")
lines(x[ix], y3[ix], col = "darkgreen")

#Leyenda
legend("bottomright", 
  legend = c("Lineal", "Grado 2", "Grado 3"), 
  lwd = 2,
  col = c("red", "blue", "darkgreen"), 
  text.col = "black"
  )
```

Se puede observar como las regresiones con polinomios de segundo y tercer grado van muy a la par, a excepción del rango de valores mayores ($x>0.65$ aproximadamente). Además, el ajuste lineal se aleja bastante de los otros dos modelos así como ocurre con los valores obtenidos para la función de coste, por lo que concluimos que en este caso, el ajuste lineal es un modelo peor.
