## Problema 1:
Considera un examen de 10 preguntas con 5 posibles respuestas para cada pregunta
en el que se responde al azar. Considera la variable aleatoria **X = número de preguntas acertadas en el examen**.

```r
# a) Calculamos la probabilidad de acertar X preguntas.
x = seq(0,10)
n = 10
p = 0.2

probs = dbinom(x, n, p)
plot(x, probs, cex = 1, pch = 16,
     xlab = "Número de preguntas acertadas",
     ylab= "Probabilidad",
     main = "Distribución de probabilidad")
```

El número de aciertos más probable es 2 preguntas.

Por otro lado, la suma de las probabilidades es:

```r
sum(probs)
```

ya que nos da la probabilidad del espacio muestral, que siempre es la unidad.

```r
# b) E(X), SD(X)
esp = n*p
std = sqrt(n*p*(1-p))
c(esp, std)
```

El valor esperado concuerda con el valor más probable para el número de aciertos que se observa en la gráfica anterior, mientras que la desviación estándar se relaciona con la anchura de la distribución.

```r
# c)
# Nueva variable aleatoria: y = 1*x - (n-x)*0.25
y = 1.25*x-2.5

# E(Y) = E(aX+b) = a*E(x)+b
espY = 1.25*esp-2.5

#Var(Y) = Var(aX+b) = a^2*Var(X)
#sd(Y) = a*sd(X)
stdY = 1.25*std
c(espY, stdY)
```

Ahora, el valor esperado es exactamente cero. Este es el motivo por el cual en los exámenes tipo test las preguntas erróneas restan puntos, para no poder obtener puntos respondiendo aleatoriamente.

## Problema 2:

Resultados de partidos de fútbol internacionales.

```r
# Leemos los datos
setwd("C:/Users/sergr/OneDrive/Universidad/MasterDataScience/NotebooksLocal")
results = read.csv("results.csv", header = TRUE)
```

```r
# a)
spain = subset(results, home_team == 'Spain')
barplot(table(spain$home_score))
```

Claramente, la distribución de los datos se asemeja a una poissoniana. De hecho, si superponemos la distribución de Poisson:

```r
barplot(table(spain$home_score)/sum(table(spain$home_score))) #Dividimos entre la suma de ocurrencias
                                                              #para normalizar y poner poisson
x = min(spain$home_score):max(spain$home_score)
y = dpois(x, mean(spain$home_score))
lines(x,y, col = "red")
```

```r
# b) Probabilidad de marcar 3 o más goles
ppois(3, mean(spain$home_score), lower.tail = F)
```

```r
# c) Prob de gol
ygol = dpois(unique(spain$home_score), mean(spain$home_score))
plot(unique(spain$home_score), ygol, pch = 16)
```

Lo más probable es que marque 2 goles:

```r
which.max(ygol)
```

Además, la suma de las probabilidades será 1, pues es la probabilidad total:

```r
sum(ygol)
```

```r
# d) E(X), SD(X)

E = mean(spain$home_score)
E
SD = sd(spain$home_score)
SD
```

$E(X)$ coincide con $\lambda$, siendo el número medio de goles marcados por partido.
$SD(X)$ se asemeja a $\lambda$ al ser una distribución poissioniana, pero observamos una ligera discrepancia, debida al limitado tamaño muestral con el que estamos trabajando. Este estimador se relacionea con la anchura de la curva de probabilidades.

## Problema 3:
El colesterol se mide en una población con media 180 mg/dL y desviación típica 25 mg/dL.

```r
# a) Prob más de 225 de colesterol
mu = 180
sd = 25

pnorm(225, mu, sd, lower.tail = F)
```

Obtenemos una probabilidad muy pequeña.

```r
# b) Percentil del 95%
qnorm(0.95, mu, sd)
```

Nos indica que por debajo de 221 se encuentran el 95% de los datos

```r
# c) Representar funciones densidad y distribución

x = seq(mu-3*sd, mu+3*sd, length = 200)
densidad = dnorm(x, mu, sd)
distribucion = pnorm(x, mu, sd)

par(mfrow = c(1,2))
plot(x, densidad, pch = 16)
plot(x, distribucion, pch = 16,, ylim = c(0,1),
       col = "blue")
```

```r
# d) E(X), SD(X)
E = mean(x)
E
SD = sd(x)
SD
```
