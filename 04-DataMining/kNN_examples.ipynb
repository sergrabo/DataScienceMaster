## Introduction to nearest-neighbour for classification
This code applies the example shown in the first slides. It uses the `iris` dataset (see `help('iris')`).

```r
data(iris)
# Target variable, iris species:
y <- iris$Species
# Predictor variables (sepal length, sepal width, petal length, petal width):
training.dataset <- iris[ ,-5]
```

```r
str(iris)
```

This is the new entry to be classified:

```r
# sepal length, sepal width, petal length, petal width
d.new <- c(5.4, 2.7, 4.6, 1.4)
```

```r
# Initialize a vector containing all distances
dists <- vector(mode = "numeric", length = nrow(iris))

# Iterate over iris instances:
for (i in 1:nrow(iris)) {
    dists[i] <- sqrt(sum(d.new - training.dataset[i,])^2)
}
```

```r
# Sort all sitamnces in ascending order
ind.sort <- sort(dists, index.return = TRUE)
y[ind.sort$ix[1]]

# The 10 nearest neighbors correspond to versicolor
y[ind.sort$ix[1:10]]
```

## Train and test datasets

```r
# train/test division 
n <- nrow(iris)
set.seed(1) # Set random seed for reproducibility of results
ind.train <- sample(1:n, floor(0.75*n))
ind.test <- setdiff(1:n, ind.train)
iris.train <- iris[ind.train, ]
iris.test <- iris[ind.test, ]
```

Here we use the k-NN implementation of package `class`. We consider just one nearest-neighbour (`k=1`)

```r
library(class)
pred <- knn(train = iris.train[ ,-5], test = iris.test[ ,-5], cl = iris.train[ ,5], k = 1)
```

```r
# Confusion matrix summarising the classification results:
table(pred, iris.test[ ,5])
```

We define a new function `accuracy`, to calculate the percentage of correctly classified cases:

```r
# Evaluation function
accuracy <- function(preds, labels) {
    stopifnot(length(preds) == length(labels))
    conf.mat <- table(preds, labels)
    sum(diag(conf.mat))/length(preds)
}
```

```r
accuracy(pred, iris.test[ ,5]) # ~95% accuracy
```

## Optimizing the number of neighbours

The package `caret` contains a user-friendly interface to simplify the estimation of the optimal number of neighbours through a user-defined error function. This is by default the accuracy for classification problems (predictand is categorical) and RMSE for regression (predictand is continuous).

```r
library(caret)
# defining hold-out cross cross-validation
set.seed(1)
trctrl <- trainControl(method = "cv", number = 2) 
# searching searching searching the optimal k
knn.fit <- train(Species ~ ., iris.train,
                 method = "knn", trControl = trctrl,
                 tuneGrid = expand.grid(k = 1:50))

plot(knn.fit)
```

Now, it is possible to perform the predictions directly with the "best" model configuration:

```r
knn.fit$best # 9 neighbours
```

```r
pred <- predict(knn.fit, iris.test)
# Accuracy of the optimal model: (~97%)
accuracy(pred, iris.test$Species)
```

## kNN for regression

For regression, we will work with the dataset `carseats` (included in the package `ISLR`). Our target variable will be `Sales`. First, we will remove all the categorical variables from the dataset, retaining only the continuous ones. We will use the function `knn.reg` from package `FNN`. As you did for the case of classification, divide the total dataset in 75% for train and 25% for test and see how the test error varies with different values of parameter `k`.


```r
library(ISLR)
attach(Carseats)
# remove categorical predictors
dataset = Carseats[, -c(7, 10, 11)] 
```

```r
# evaluation function, Root Mean Square Error
rmse <- function(x, y) {
    sqrt(mean((x - y)^2))
}
```

```r
# train/test dataset partition
n <- nrow(dataset)
set.seed(1)
indtrain <- sample(1:n, round(0.75*n))
dataset.train <- dataset[indtrain, ]
indtest <- setdiff(1:n, indtrain)
dataset.test <- dataset[indtest, ]
```

In this example the search for an optimal `k` is programmed "by hand":

```r
# test error as a function of k
library(FNN)
kmax <- 50
test.err <- vector(mode = "numeric", length = kmax)
for (k in 1:kmax) {
    pred = knn.reg(train = dataset.train[,-1],
                   test = dataset.test[,-1],
                   y = dataset.train$Sales, k = k)
    test.err[k] = rmse(pred$pred, as.numeric(dataset.test$Sales))
}
plot(1:kmax, test.err, type = "o", pch = 19, xlab = "k", ylab = "RMSE")
grid()
```

## Rescaling the predictors

In the following example we assess the effect of a proper rescaling of the predictors, by comparing the results with the previous example. In the first plot, the distributional differences among predictors is shown as a boxplot:

```r
boxplot(dataset[,-1])

# test error as a function of k (for standardized data)
test.err2 <- vector(mode = "numeric", length = kmax)
for (k in 1:kmax) {
    pred = knn.reg(train = scale(dataset.train[,-1]),
                   test = scale(dataset.test[,-1]),
                   y = dataset.train$Sales, k = k)
    test.err2[k] = rmse(pred$pred, as.numeric(dataset.test$Sales))
}
```

The optimization results in terms of RMSE are here compared with the previous raw-predictor version:

```r
plot(1:kmax, test.err, type = "o", pch = 19, xlab = "k", ylab = "RMSE", ylim = c(2.3,3.3))
lines(1:kmax, test.err2, type = "o", pch = 19, col = "red")
grid()
legend("topright", title = "predictor preprocess:",
       c("without scaling", "with scaling"), 
       lty = 1, col = c("black", "red"))
```
